```python
x = df.iloc[:, [ 0,1, 2, 3]].values
x
```

```python
y = df.iloc[:, [-1]].values
y 
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the results onto a line graph, 
# `allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss,marker="o")
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

From the above Visulazation of "The Elbow Method", WSCC(Within cluster sum of square) doesn't decrease significantly with every iteration.
So, we choose Number of cluster 3

# KMeans Clustering

```python
kmeans=KMeans(n_clusters=3,init = 'k-means++',max_iter = 300, n_init = 10,random_state=0)
y_means=kmeans.fit_predict(x)
print("The predicted clusters are:\n",y_means)
print("The centers are:\n",kmeans.cluster_centers_)
```

```python
pd.DataFrame({'Original':y, 'Predicted': y_means})
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
conf_mat= confusion_matrix(y,y_means)
print("The confusion matrix :\n",conf_mat)
ac_score= accuracy_score(y,y_means)
print("The accuracy score:",ac_score)
```

```python
# Visualising the clusters - On the first two columns
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
plt.show()
```
